## **체육복**

전체 학생의 수 n, 체육복을 도난당한 학생들의 번호가 담긴 배열 lost, 여벌의 체육복을 가져온 학생들의 번호가 담긴 배열 reserve가 매개변수로 주어질 때, 체육수업을 들을 수 있는 학생의 최댓값을 return 하도록 solution 함수를 작성해주세요. 바로 앞번호의 학생이나 바로 뒷번호의 학생에게만 체육복을 빌려줄 수 있습니다.
* ex) n = 5, lost = [2, 4], reserve = [1, 3, 5] -> 5

```python
#try1 : score 70/100
def solution(n, lost, reserve):
    
    #체육복 소지 개수 리스트 만들기
    
    lst = [1] * n
    for i in range(n):
        if i+1 in lost:
            lst[i] -= 1
        if i+1 in reserve:
            lst[i] += 1
    
    # 2, 0을 1, 1로 만든 후
    # 0, 2를 1, 1로 만들어 result1 도출
    
        
    lst1 = lst.copy()        
    for i in range(0, n-1):
        if lst1[i] == 2 and lst1[i+1] == 0:
            lst1[i] = 1
            lst1[i+1] = 1
            
    for i in range(0, n-1):
        if lst1[i] == 0 and lst1[i+1] == 2:
            lst1[i] = 1
            lst1[i+1] = 1
    
    result1 = sum(lst1)-lst1.count(2)

    
    # 0, 2을 1, 1로 만든 후
    # 2, 0를 1, 1로 만들어 result2 도출
        
        
    lst2 = lst.copy()      
    for i in range(0, n-1):
        if lst2[i] == 0 and lst2[i+1] == 2:
            lst2[i] = 1
            lst2[i+1] = 1
            
    for i in range(0, n-1):
        if lst2[i] == 2 and lst2[i+1] == 0:
            lst2[i] = 1
            lst2[i+1] = 1
            
    result2 = sum(lst2)-lst2.count(2)
    
    
    
    #n보다 같거나 작은 값 중 제일 큰 값 도출
    #마지막 값은 그냥 직관적으로 추측해서 넣었음
    
    return min(max(result1, result2, n - len(lost) + len(reserve)), n)
```

```python
#try3 score 100/100
def solution(n, lost, reserve):
    
    #체육복 소지 개수 리스트 만들기
    
    lst = [1] * n
    for i in range(n):
        if i+1 in lost:
            lst[i] -= 1
        if i+1 in reserve:
            lst[i] += 1
    dic = {}
    for i in range(len(lst)):
        dic[i+1] = lst[i]
    
    # 번호를 기준으로 도출
    
    for key in dic.keys():
        if dic[key] == 0:
            if key-1 > 0 and dic[key-1] == 2:
                dic[key] = 1
                dic[key-1] = 1
            elif key+1 <= len(dic) and dic[key+1] == 2:
                dic[key] = 1
                dic[key+1] = 1

    return(sum(i if i <= 1 else 1 for i in dic.values()))
```

* #### **논리적으로 순차적으로 생각하기.**
맨 앞 사람은 항상 뒷사람에게만 빌려줄 수 있다. 같은 방식을 반복하면 일렬로 수행할 수 있다.

```python
#answer
def solution(n, lost, reserve):
    
    #처음부터 학생번호를 기준으로 그룹을 나눴다.
    
    set_lost = set(lost) - set(reserve) 
    
    set_reserve = set(reserve) - set(lost) 
    
    #빌려주는 사람의 관점에서 식 구축.
    
    for i in set_reserve: 
        if i-1 in set_lost: 
            set_lost.remove(i-1) 
        elif i+1 in set_lost: 
            set_lost.remove(i+1) 
    return n -len(set_lost)
```
